```python
data_mahasiswa = []

def validasi_nim(nim):
    return nim.isdigit() and len(nim) == 11

def validasi_nama(nama):
    return len(nama) >= 6

def validasi_alamat(alamat):
    return len(alamat) >= 10

def validasi_kelas(kelas):
    return kelas in ["TI24H", "TI24A", "TI24G", "TI24F"]

# Fungsi tambah data
def tambah_mahasiswa():
    nim = input("Masukkan NIM (11 digit): ")
    if not validasi_nim(nim):
        print("NIM harus 11 digit angka.")
        return

    nama = input("Masukkan Nama (minimal 6 karakter): ")
    if not validasi_nama(nama):
        print("Nama minimal 6 karakter.")
        return

    alamat = input("Masukkan Alamat (minimal 10 karakter): ")
    if not validasi_alamat(alamat):
        print("Alamat minimal 10 karakter.")
        return

    kelas = input("Masukkan Kelas (TI24H, TI24A, TI24G, TI24F): ")
    if not validasi_kelas(kelas):
        print("Kelas tidak valid.")
        return

    mahasiswa = {
        "NIM": nim,
        "Nama": nama,
        "Alamat": alamat,
        "Kelas": kelas
    }
    data_mahasiswa.append(mahasiswa)
    print("Data mahasiswa berhasil ditambahkan.")

# Fungsi tampil data
def tampilkan_data():
    if not data_mahasiswa:
        print("Belum ada data mahasiswa.")
        return
    print("Daftar Mahasiswa:")
    for idx, mhs in enumerate(data_mahasiswa, start=1):
        print(f"{idx}. NIM: {mhs['NIM']}, Nama: {mhs['Nama']}, Alamat: {mhs['Alamat']}, Kelas: {mhs['Kelas']}")


def menu():
    while True:
        print("\n====== MENU MAHASISWA ======")
        print("1. Tambah Data Mahasiswa")
        print("2. Tampilkan Data Mahasiswa")
        print("3. Keluar")
        pilihan = input("Pilih menu (1/2/3): ")

        if pilihan == "1":
            tambah_mahasiswa()
        elif pilihan == "2":
            tampilkan_data()
        elif pilihan == "3":
            print("Keluar dari program.")
            break
        else:
            print("Pilihan tidak valid. Coba lagi.")


menu()
```
