```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")
plt.style.use("ggplot")

data = {
    'Country': ['Indonesia', 'USA', 'China', 'India', 'Brazil', 'Russia', 'Nigeria', 'Japan', 'Germany', 'UK'],
    'IQ': [87, 98, 105, 82, 87, 96, 84, 106, 102, 100]
}

df = pd.DataFrame(data)

plt.figure(figsize=(10, 6))
sns.barplot(x='IQ', y='Country', data=df, palette='mako')
plt.title('IQ Rata-Rata per Negara (Bar Chart)')
plt.xlabel('IQ')
plt.ylabel('Negara')
plt.tight_layout()
plt.show()

plt.figure(figsize=(8, 8))
plt.pie(df['IQ'], labels=df['Country'], autopct='%1.1f%%', startangle=140, colors=sns.color_palette('mako', len(df)))
plt.title('Distribusi IQ Rata-Rata Negara (Pie Chart)')
plt.axis('equal')
plt.tight_layout()
plt.show()

plt.figure(figsize=(10, 6))
sns.lineplot(x='Country', y='IQ', data=df, marker='o', linewidth=2)
plt.title('IQ Rata-Rata per Negara (Line Chart)')
plt.xlabel('Negara')
plt.ylabel('IQ')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
